```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Load data
commits = pd.read_csv("data/github_commits.csv")
messages = pd.read_csv("data/messages.csv")
tickets = pd.read_csv("data/tickets.csv")

# Merge all data by date
df = pd.merge(commits, messages, on="date")
df = pd.merge(df, tickets, on="date")

# Convert date column
df['date'] = pd.to_datetime(df['date'])

# Sort
df = df.sort_values("date")

df.head()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Segoe UI Emoji'  # or Arial Unicode MS (on Mac)

plt.figure(figsize=(10, 4))
plt.plot(df['date'], df['commits'], marker='o')
plt.title("📈 Commits Over Time")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(data=df, x='messages', y='commits')
plt.title("🧠 Commits vs Messages")
plt.grid(True)
plt.show()
```

```python



# print(df.isnull().sum())
df['commits'] = pd.to_numeric(df['commits'], errors='coerce')
df['messages'] = pd.to_numeric(df['messages'], errors='coerce')
df['tickets_closed'] = pd.to_numeric(df['tickets_closed'], errors='coerce')

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("🔥 Correlation Matrix")
plt.show()


print(df[['commits', 'messages', 'tickets_closed']])

```
